```python
# many-to-one
df1 = pd.DataFrame({'patient': ['Bob', 'Lisa', 'Sue'], 'repas': ['SS', 'SS', 'SSR']})
df2 = pd.DataFrame({'repas': ['SS', 'SSR'], 'explication': ['sans sel', 'sans sucre']})
print(df1)
print(df2)
pd.merge(df1,df2)
```

```python
# many-to-many
df1 = pd.DataFrame({'patient': ['Bob', 'Lisa', 'Sue'], 'repas': ['SS', 'SS', 'SSR']})
df2 = pd.DataFrame({'repas': ['SS', 'SS', 'SSR'], 'explication': ['sans sel', 'lÃ©gumes', 'sans sucre']})
print(df1)
print(df2)
pd.merge(df1,df2)
```

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Lisa', 'Sue'], 'group': ['Accounting', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Sue'], 'hire_date': [2004, 2008, 2014]})
print(df1)
print(df2)
pd.merge(df1,df2, on='employee')
```
